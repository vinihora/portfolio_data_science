### Dicionário de Variáveis
- ```iso_code``` - ISO 3166 - Código de 3 letras que representam os países
- ```continent``` - Continente o qual o país se localiza
- ```location``` - Localização geográfica
- ```date``` - Data da observação
- ```total_cases``` - Total confirmado de casos de COVID-19, a depender da fonte, pode incluir casos prováveis
- ```new_cases``` - Novos casos confirmados
- ```new_cases_smoothed``` - Novos casos (Média dos últimos 7 dias), incluindo o dia atual
- ```total_deaths``` - Número total de mortes pelo COVID-19
- ```new_deaths``` - Novas mortes pelo COVID-19
- ```new_deaths_smoothed``` - Novas mortes pelo COVID-10 (Média dos últimos 7 dias), incluindo o dia atual
- ```total_cases_per_million``` - Número total de casos a cada 1.000.000 de pessoas
- ```new_cases_per_million``` - Novos casos de COVID-19 a cada 1.000.000 de pessoas
- ```new_cases_smoothed_per_million``` - Novos casos nos últimos 7 dias
- ```total_deaths_per_million``` - Número total de mortes a cada 1.000.000 de pessoas
- ```new_deaths_per_million``` - Novas mortes confirmadas a cada 1.000.000 de pessoas
- ```new_deaths_smoothed_per_million``` - Média de novas mortes nos últimos 7 dias
- ```reproduction_rate``` - Taxa de reprodução do COVID-19 (estimativa)
- ```icu_patients``` e ```icu_patients_per_million``` - Número de pacientes em UTIs
- ```hosp_patients```e ```hosp_patients_Per_million``` - Número de pacientes com COVID-19 nos hospitais no dia observado
- ```weekly_icu_admissions``` e ```weekly_icu_admissions_per_million``` - Número de pacientes entrantes em UTIs na última semana (dia atual + 6 dias anteriores)
- ```weekly_hosp_admissions``` e ```weekly_hosp_admissions_per_million``` - Número de pacientes entranes em hospitais na última semana (dia atual + 6 dias anteriores)
- ```total_tests``` - Número total de testes
- ```new_tests``` - Número de novos testes no dia
- ```total_tests_per_thousand``` - Número total de testes a cada 1000 pessoas
- ```new_tests_per_thousand```- Número de novos testes a cada 1000 pessoas
- ```new_tests_smoothed``` e ```new_tests_smoothed_per_thousand``` - Média dos últimos 7 dias de novos testes
- ```positive_rate``` - A porcentagem de teste de COVID-19 que são positivos (Média dos últimos 7 dias)
- ```tests_per_case``` - A cada quantos testes é detectado um caso de COVID-19 confirmado
- ```tests_units``` - Número de locais que disponibilizam os testes para a COVID-19, que forneceram os dados para a base
- ```total_vaccinations``` e ```total_vaccinations_per_hundred``` - Número total de vacinas aplicadas no país
- ```people_vaccinated``` e ```people_vaccinated_per_hundred``` - Número total de pessoas que receberam pelo menos uma dose
- ```people_fully_vaccinated``` e ```people_fully_vaccinated_per_hundred``` - Número total de pessoas que receberam todas as doses pre-escritas pelo protocólo inicial de vacinação
- ```total_boosters``` e ```total_boosters_per_hundred``` - Número total de doses de reforço administradas
- ```new_vaccinations``` , ```new_vaccinations_smoothed``` e ```new_vaccinations_smoothed_per_million``` - Novas doses administradas
- ```stringency_index``` - Resposta do governo a restrição (fechamento de escolas, escritórios, viajens e etc), variando de 0 a 100, 100 = maior restrição
- ```population_density``` - Número de pessoas, divido pela extensão territorial (medido em km²)
- ```median_age``` - Média da idade da população
- ```aged_65_older``` - Porcentagem da população a cima de 65 anos
- ```aged_70_older``` - Porcentagem da população a cima de 70 anos
- ```gdp_per_capita``` - Produto Interno Bruto (PIB) per Capita em dólares
- ```extreme_poverty``` - Porcentagem da população em extrema pobreza
- ```cardiovasc_death_rate``` - Porcentagem de mortes decorrentes de doenças cardio-vasculares em 2017 - (Número anual de mortes a cada 100.000 pessoas)
- ```diabetes_prevalence``` - Porcentagem da população com diabetes (dos 20 aos 79) em 2017
- ```female_smokers``` - Porcentagem de mulheres que fumam
- ```male_smokers``` - Porcentagem de homens que fumam
- ```handwashing_facilities``` - Porcentagem da população com saneamento básico
- ```hospital_beds_per_thousand``` - Camas de hospital a cada 1000 pessoas
- ```life_expectancy``` - Expectativa de vida
- ```human_development_index``` - Índice de desenvolvimento humano (Varia de 0 a 1, sendo 1 = máximo), analisa 3 pilares de um país: expectativa/qualidade de vida, Educação e qualidade de vida
- ```population``` - População atual do país
- ```excess_mortality_cumulative_absolute``` - Diferença absoluta entre o número real de mortes ocorridas e a previsão da mesma data de n° de mortos no ano anterior
- ```excess_mortality_cumulative``` e ```excess_mortality_cumulative_per_million``` - Diferença em porcentagem entre o número real de mortes ocorridas e a previsão da mesma data de n° de mortos no ano anterior

````
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# importar o dataset COVID-19 para um DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/vinihora/portfolio_data_science/main/dados/owid-covid-data.csv', parse_dates=True)
````

## Classificando os países entre desenvolvidos e subdesenvolvidos

- A fim classificar os países entre os 2 diferentes grupos, adotaremos algumas premissas para considerá-los como desenvolvidos ou subdesenvolvidos a fim de atender a necessidade do estudo

- A primeira premissa utilizada no modelo será em relação ao **Índice de Desenvolvimento Humano (IDH)**, que está relacionado a variável ```humam_development_index```. Os países desenvolvidos terão de possuir um valor **acima de 0.9**

- Outra premissa utilizada será o PIB per capita superior a $30.000 por ano para ser considerado um país desenvolvido

## Aplicando as premissas
- Primeiro, vamos descobrir quais países atendem aos requisitos, com o código de bloco abaixo

````
# Localizando no DataFrame países que possuem um IDH > 0.9 e PIB per capita > 300000
developed_countries = df.loc[(df['human_development_index'] > 0.9) & (df['gdp_per_capita'] > 30000)].location.unique()
developed_countries
````

- No próximo bloco de código será adicionado uma nova coluna ao data frame com o nome ```developed```, sinalizando se o país está na lista de desenvolvidos ou não

````
# Criando a coluna, atribuindo todos os valores como False
df['developed'] = False

# Atribuindo o valor True para os países que se enquadram nas premissas adotadas
df.loc[(df['human_development_index'] > 0.9) & (df['gdp_per_capita'] > 30000), 'developed'] = True
````

- Verificado se a adição foi feita com sucesso

````
# Verificando as 5 primeiras observações de países desenvolvidos
df.loc[(df['developed'] == True)].head(5)
````

````
# Verificando as 5 primeiras observações de países não desenvolvidos
df.loc[(df['developed'] == False)].head(5)
````

- A partir das amostrar acima, podemos verificar que a nova coluna foi adicionada ao DataFrame e está de acordo com as regras pré-estabelecidas
