### Dicionário de Variáveis
- ```iso_code``` - ISO 3166 - Código de 3 letras que representam os países
- ```continent``` - Continente o qual o país se localiza
- ```location``` - Localização geográfica
- ```date``` - Data da observação
- ```total_cases``` - Total confirmado de casos de COVID-19, a depender da fonte, pode incluir casos prováveis
- ```new_cases``` - Novos casos confirmados
- ```new_cases_smoothed``` - Novos casos (Média dos últimos 7 dias), incluindo o dia atual
- ```total_deaths``` - Número total de mortes pelo COVID-19
- ```new_deaths``` - Novas mortes pelo COVID-19
- ```new_deaths_smoothed``` - Novas mortes pelo COVID-10 (Média dos últimos 7 dias), incluindo o dia atual
- ```total_cases_per_million``` - Número total de casos a cada 1.000.000 de pessoas
- ```new_cases_per_million``` - Novos casos de COVID-19 a cada 1.000.000 de pessoas
- ```new_cases_smoothed_per_million``` - Novos casos nos últimos 7 dias
- ```total_deaths_per_million``` - Número total de mortes a cada 1.000.000 de pessoas
- ```new_deaths_per_million``` - Novas mortes confirmadas a cada 1.000.000 de pessoas
- ```new_deaths_smoothed_per_million``` - Média de novas mortes nos últimos 7 dias
- ```reproduction_rate``` - Taxa de reprodução do COVID-19 (estimativa)
- ```icu_patients``` e ```icu_patients_per_million``` - Número de pacientes em UTIs
- ```hosp_patients```e ```hosp_patients_Per_million``` - Número de pacientes com COVID-19 nos hospitais no dia observado
- ```weekly_icu_admissions``` e ```weekly_icu_admissions_per_million``` - Número de pacientes entrantes em UTIs na última semana (dia atual + 6 dias anteriores)
- ```weekly_hosp_admissions``` e ```weekly_hosp_admissions_per_million``` - Número de pacientes entranes em hospitais na última semana (dia atual + 6 dias anteriores)
- ```total_tests``` - Número total de testes
- ```new_tests``` - Número de novos testes no dia
- ```total_tests_per_thousand``` - Número total de testes a cada 1000 pessoas
- ```new_tests_per_thousand```- Número de novos testes a cada 1000 pessoas
- ```new_tests_smoothed``` e ```new_tests_smoothed_per_thousand``` - Média dos últimos 7 dias de novos testes
- ```positive_rate``` - A porcentagem de teste de COVID-19 que são positivos (Média dos últimos 7 dias)
- ```tests_per_case``` - A cada quantos testes é detectado um caso de COVID-19 confirmado
- ```tests_units``` - Número de locais que disponibilizam os testes para a COVID-19, que forneceram os dados para a base
- ```total_vaccinations``` e ```total_vaccinations_per_hundred``` - Número total de vacinas aplicadas no país
- ```people_vaccinated``` e ```people_vaccinated_per_hundred``` - Número total de pessoas que receberam pelo menos uma dose
- ```people_fully_vaccinated``` e ```people_fully_vaccinated_per_hundred``` - Número total de pessoas que receberam todas as doses pre-escritas pelo protocólo inicial de vacinação
- ```total_boosters``` e ```total_boosters_per_hundred``` - Número total de doses de reforço administradas
- ```new_vaccinations``` , ```new_vaccinations_smoothed``` e ```new_vaccinations_smoothed_per_million``` - Novas doses administradas
- ```stringency_index``` - Resposta do governo a restrição (fechamento de escolas, escritórios, viajens e etc), variando de 0 a 100, 100 = maior restrição
- ```population_density``` - Número de pessoas, divido pela extensão territorial (medido em km²)
- ```median_age``` - Média da idade da população
- ```aged_65_older``` - Porcentagem da população a cima de 65 anos
- ```aged_70_older``` - Porcentagem da população a cima de 70 anos
- ```gdp_per_capita``` - Produto Interno Bruto (PIB) per Capita em dólares
- ```extreme_poverty``` - Porcentagem da população em extrema pobreza
- ```cardiovasc_death_rate``` - Porcentagem de mortes decorrentes de doenças cardio-vasculares em 2017 - (Número anual de mortes a cada 100.000 pessoas)
- ```diabetes_prevalence``` - Porcentagem da população com diabetes (dos 20 aos 79) em 2017
- ```female_smokers``` - Porcentagem de mulheres que fumam
- ```male_smokers``` - Porcentagem de homens que fumam
- ```handwashing_facilities``` - Porcentagem da população com saneamento básico
- ```hospital_beds_per_thousand``` - Camas de hospital a cada 1000 pessoas
- ```life_expectancy``` - Expectativa de vida
- ```human_development_index``` - Índice de desenvolvimento humano (Varia de 0 a 1, sendo 1 = máximo), analisa 3 pilares de um país: expectativa/qualidade de vida, Educação e qualidade de vida
- ```population``` - População atual do país
- ```excess_mortality_cumulative_absolute``` - Diferença absoluta entre o número real de mortes ocorridas e a previsão da mesma data de n° de mortos no ano anterior
- ```excess_mortality_cumulative``` e ```excess_mortality_cumulative_per_million``` - Diferença em porcentagem entre o número real de mortes ocorridas e a previsão da mesma data de n° de mortos no ano anterior

````
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math


# importar o dataset COVID-19 para um DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/vinihora/portfolio_data_science/main/dados/owid-covid-data.csv', parse_dates=True)
````

## Classificando os países entre desenvolvidos e subdesenvolvidos

- A fim classificar os países entre os 2 diferentes grupos, adotaremos algumas premissas para considerá-los como desenvolvidos ou subdesenvolvidos a fim de atender a necessidade do estudo

- A primeira premissa utilizada no modelo será em relação ao **Índice de Desenvolvimento Humano (IDH)**, que está relacionado a variável ```humam_development_index```. Os países desenvolvidos terão de possuir um valor **acima de 0.9**

- Outra premissa utilizada será o PIB per capita superior a $30.000 por ano para ser considerado um país desenvolvido

## Aplicando as premissas
- Primeiro, vamos descobrir quais países atendem aos requisitos, com o código de bloco abaixo

````
# Localizando no DataFrame países que possuem um IDH > 0.9 e PIB per capita > 300000
developed_countries = df.loc[(df['human_development_index'] > 0.9) & (df['gdp_per_capita'] > 30000)].location.unique()
developed_countries
````

- No próximo bloco de código será adicionado uma nova coluna ao data frame com o nome ```developed```, sinalizando se o país está na lista de desenvolvidos ou não

````
# Criando a coluna, atribuindo todos os valores como False
df['developed'] = False

# Atribuindo o valor True para os países que se enquadram nas premissas adotadas
df.loc[(df['human_development_index'] > 0.9) & (df['gdp_per_capita'] > 30000), 'developed'] = True
````

- Verificado se a adição foi feita com sucesso

````
# Verificando as 5 primeiras observações de países desenvolvidos
df.loc[(df['developed'] == True)].head(5)
````

````
# Verificando as 5 primeiras observações de países não desenvolvidos
df.loc[(df['developed'] == False)].head(5)
````

- A partir das amostrar acima, podemos verificar que a nova coluna foi adicionada ao DataFrame e está de acordo com as regras pré-estabelecidas

- A fim de reduzir a possibilidade de dados faltantes, comprometendo a análise, os grupos de países desenvolvidos será reduzido.

## **2.4 Brasil X Países desenvolvidos**

- A fim de colocar os dados em uma linha do tempo do início da pandemia, até os dias atuais, selecionaremos todas as datas disponíveis no dataset.

- Além disso, criaremos um novo DataFrame somente com os dados brasileiros, a fim de facilitar a visualização e comparação com os desenvolvidos.

````
# Seleciona as datas únicas disponíveis no dataset
unique_days = df_change['date'].unique()

# Cria um novo DataFrame somente com as informações relacionadas ao Brasil
df_brasil = df.loc[df['location'] == "Brazil"]

# Transforma os dados do formato string para o formato date
df_brasil.date = pd.to_datetime(df_brasil.date)
````

## **Países desenvolvidos**

- Os gráficos a seguir são construções que levam em consideração a média de todos os países selecionados como desenvolvidos anteriormente.
- Em cada uma das datas selecionados a pouco, foi calculada a média dos países para cada índice, e a partir disso foram adicionadas novas informações a nossa base de dados, que agrupa os países desenvolvidos selecionados, que serão utilizados na comparação com o Brasil.

### **Vacinas aplicadas**
- A partir do gráfico abaixo, podemos confirmar a suspeita que foi apresentada anteriormente. Apesar do atraso inicial na administração de vacinas, o país logo se recuperou e a partir dai manteve uma trajetória equivalente em relação a média dos países desenvolvidos.
- Apesar de não se apresentar superior em nenhum perído, pode-se dizer que após a recuperação, o Brasil conseguiu apresentar um bom desempenho na aplicação de vacinas ao longo do tempo, levando em conta o desempenho dos países desenvolvidos.

````
# Criação do gráfico
fig_vaccination, ax_vaccination = plt.subplots(1,1, figsize=(12,3))

# Criando novo DataFrame com o total de vacinas a cada 100 pessoas (média dos países desenvolvidos)
new_data = {'day': [], 'total_vaccination_per_hundred': []}
df_developed_vaccination = pd.DataFrame(new_data)

# Loop em cada dia do dataset
for day in unique_days:
  # Filtra o dia e verifica se o país está classificado como desenvolvido ou não
  actual_day_df = df_change.loc[(df_change['date'] == day) & (df_change['developed'] == True)]

  # verifica se o valor não é nulo
  if not math.isnan(actual_day_df.total_vaccinations_per_hundred.mean()):
    # Adiciona uma nova row no Dataset
    if len(df_developed_vaccination) > 0:
      if df_developed_vaccination.iloc[-1].total_vaccination_per_hundred <= actual_day_df.total_vaccinations_per_hundred.mean():
        new_row = {'day': day, 'total_vaccination_per_hundred': actual_day_df.total_vaccinations_per_hundred.mean()}
        df_developed_vaccination.loc[len(df_developed_vaccination)] = new_row
    elif len(df_developed_vaccination) == 0:
        new_row = {'day': day, 'total_vaccination_per_hundred': actual_day_df.total_vaccinations_per_hundred.mean()}
        df_developed_vaccination.loc[len(df_developed_vaccination)] = new_row

# Transforma os dados de string para Date
df_developed_vaccination.day = pd.to_datetime(df_developed_vaccination.day)

# Plota o gráfico na tela (Desenvolvidos X Brasil)
ax_vaccination.plot(df_developed_vaccination['day'], df_developed_vaccination['total_vaccination_per_hundred'], label='Desenvolvidos')
ax_vaccination.plot(df_brasil['date'], df_brasil['total_vaccinations_per_hundred'], label='Brasil')
ax_vaccination.set_title('Total de vacinas aplicadas a cada 100 pessoas')
ax_vaccination.legend()
````

### **Doses adicionais**
- O gráfico revela um enorme atraso do Brasil em relação aos países desenvolvidos.
- Existiram alguns pontos em que o país ensaia uma boa recuperação com o objetivo de acompanhar os níveis nos países desenvolvidos, contudo isso logo se enfraquece e o Brasil volta a uma constância no quesito de aplicação de doses adicionais, assim como ocorreu durante os períodos de julho e agosto de 2022, que foram os que mais se aproximaram dos níveis de países desenvolvidos.
- Além disso, pode-se perceber um crescimento no distanciamento das duas curvas a partir de setembro de 2022. Enquanto os países desenvolvidos mostraram uma preocupação muito grande em aumentar o número de doses de reforços, o Brasil manteve um nível constante, muito por conta do "relaxamento" após a grande redução de novos casos confirmados no país.

````
fig_boosters, ax_boosters = plt.subplots(1,1, figsize=(12,3))

new_data = {'day': [], 'total_boosters_per_hundred': []}
df_developed_boosters = pd.DataFrame(new_data)

for day in unique_days:
  actual_day_df = df_change.loc[(df_change['date'] == day) & (df_change['developed'] == True)]
  if not math.isnan(actual_day_df.total_boosters_per_hundred.mean()):
    if len(df_developed_boosters) > 0:
      if df_developed_boosters.iloc[-1].total_boosters_per_hundred <= actual_day_df.total_boosters_per_hundred.mean():
        new_row = {'day': day, 'total_boosters_per_hundred': actual_day_df.total_boosters_per_hundred.mean()}
        df_developed_boosters.loc[len(df_developed_boosters)] = new_row
    elif len(df_developed_boosters) == 0:
        new_row = {'day': day, 'total_boosters_per_hundred': actual_day_df.total_boosters_per_hundred.mean()}
        df_developed_boosters.loc[len(df_developed_boosters)] = new_row

df_developed_boosters.day = pd.to_datetime(df_developed_boosters.day)

ax_boosters.plot(df_developed_boosters['day'], df_developed_boosters['total_boosters_per_hundred'], label='Desenvolvidos')
ax_boosters.plot(df_brasil['date'], df_brasil['total_boosters_per_hundred'], label='Brasil')
ax_boosters.set_title('Total de doses adicionais aplicadas a cada 100 pessoas')
ax_boosters.legend()
````

### **Testes aplicados**
- Nessa gráfico o descolamento das curvas dos testes aplicados fica ainda mais evidente. Enquanto os países desenvolvidos apresentaram um nível maior de testes aplicados do que habitantes, o Brasil teve um desempenho muito inferior

````
fig_tests, ax_tests = plt.subplots(1,1, figsize=(12,3))

new_data = {'day': [], 'total_tests_per_thousand': []}
df_developed_tests = pd.DataFrame(new_data)

for day in unique_days:
  actual_day_df = df_change.loc[(df_change['date'] == day) & (df_change['developed'] == True)]
  if not math.isnan(actual_day_df.total_tests_per_thousand.mean()):
    if len(df_developed_tests) > 0:
      if df_developed_tests.iloc[-1].total_tests_per_thousand <= actual_day_df.total_tests_per_thousand.mean():
        new_row = {'day': day, 'total_tests_per_thousand': actual_day_df.total_tests_per_thousand.mean()}
        df_developed_tests.loc[len(df_developed_tests)] = new_row
    elif len(df_developed_tests) == 0:
        new_row = {'day': day, 'total_tests_per_thousand': actual_day_df.total_tests_per_thousand.mean()}
        df_developed_tests.loc[len(df_developed_tests)] = new_row

df_developed_tests.day = pd.to_datetime(df_developed_tests.day)

ax_tests.plot(df_developed_tests['day'], df_developed_tests['total_tests_per_thousand'], label='Desenvolvidos')
ax_tests.plot(df_brasil['date'], df_brasil['total_tests_per_thousand'], label='Brasil')
ax_tests.set_title('Total de testes aplicadaos a cada 100 pessoas')
ax_tests.legend()
````

### **Níveis de restrições**
- O nível de restrição do Brasil durante o período da pandemia, foi um assunto extremamente delicado devido ao conturbado cenário político presente no país.
- Contudo, é possível perceber que ao contrário do que muitos pensavam em relação a pandemia, o Brasil se manteve muito mais restrito durante o período pandêmico, principalmente no início da pandemia, do que os países desenvolvidos.
- Fato este que pode ter sido um fator que suavizou o número de casos no Brasil, sendo o responsável por levar a média de casos confirmados no país bara baixo, em relação aos desenvolvidos.

````
fig_stringency, ax_stringency = plt.subplots(1,1, figsize=(12,3))

new_data = {'day': [], 'stringency_index': []}
df_developed_stringency = pd.DataFrame(new_data)

for day in unique_days:
  actual_day_df = df_change.loc[(df_change['date'] == day) & (df_change['developed'] == True)]
  new_row = {'day': day, 'stringency_index': actual_day_df.stringency_index.mean()}
  df_developed_stringency.loc[len(df_developed_stringency)] = new_row

df_developed_stringency.day = pd.to_datetime(df_developed_stringency.day)
df_developed_stringency.dropna(subset=['stringency_index'], inplace=True)

ax_stringency.plot(df_developed_stringency['day'], df_developed_stringency['stringency_index'], label='Desenvolvidos')
ax_stringency.plot(df_brasil['date'], df_brasil['stringency_index'], label='Brasil')
ax_stringency.set_title('Indíce de restrições adotadas pelos governos')
ax_stringency.legend()
````

### **Novos casos confirmados**
- Este é um dos gráfico de maior fragilidade em toda análise.
- Apesar do número menor de casos confirmados no país em relação aos países desenvolvidos, outras informações analisadas nos conduzem a acreditar que esse número menor de casos revelados diariamente no país não reflete a realidade.
- Existem 2 principais razões pela qual isso ocorre:
1. O número de testes realizados no país foi infinitamente menor se comparado aos países desenolvidos, o que afeta diretamente o número de casos confirmados no país, visto que em casos de infecções com sintomas leves ou assintomáticos, não foram inclusos no histórico brasileiro.
2. No último gráfico da sessão, é possível notar um número maior de mortes confirmadas no Brasil em relação ao resto do mundo na maior parte dos períodos. Ou seja, levando em conta que a letalidade do vírus nas diferentes regiões é relativamente semelhante, o número de casos existentes no Brasil deve ser equivalente ou até maior do que nos países desenvolvidos, sendo mascarado pelo baixo número de testes aplicados.

````
fig_new_cases, ax_new_cases = plt.subplots(1,1, figsize=(12,3))

new_data = {'day': [], 'new_cases_smoothed_per_million': []}
df_developed_cases = pd.DataFrame(new_data)

for day in unique_days:
  actual_day_df = df_change.loc[(df_change['date'] == day) & (df_change['developed'] == True)]
  new_row = {'day': day, 'new_cases_smoothed_per_million': actual_day_df.new_cases_smoothed_per_million.mean()}
  df_developed_cases.loc[len(df_developed_cases)] = new_row

df_developed_cases.day = pd.to_datetime(df_developed_cases.day)

ax_new_cases.plot(df_developed_cases['day'], df_developed_cases['new_cases_smoothed_per_million'], label='Desenvolvidos')
ax_new_cases.plot(df_brasil['date'], df_brasil['new_cases_smoothed_per_million'], label='Brasil')
ax_new_cases.set_title('Novos casos a cada 1 milhão de pessoas')
ax_new_cases.legend()
````

### **Mortes confirmadas**
- Podemos observas que o país sofreu grandes picos de mortes em relação aos países desenvolvidos, principalmente no início da pandemia no qual a população não estava devidamente vacina.
- Esse número alavancado de mortes no início da pandemia, pode ser decorrente da infraestutura brasileira, que provavelmente não suportou o número crescente de internações por conta da COVID-19 e pessoas que poderiam se recuperar da doença, devido a diferença estrutural do Brasil com países desenvolvidos, não foi possível suportar um crescente número de casos.

````
fig_new_deaths, ax_new_deaths = plt.subplots(1,1, figsize=(12,3))

new_data = {'day': [], 'new_deaths_smoothed_per_million': []}
df_developed_deaths = pd.DataFrame(new_data)

for day in unique_days:
  actual_day_df = df_change.loc[(df_change['date'] == day) & (df_change['developed'] == True)]
  new_row = {'day': day, 'new_deaths_smoothed_per_million': actual_day_df.new_deaths_smoothed_per_million.mean()}
  df_developed_deaths.loc[len(df_developed_deaths)] = new_row

df_developed_deaths.day = pd.to_datetime(df_developed_deaths.day)

ax_new_deaths.plot(df_developed_deaths['day'], df_developed_deaths['new_deaths_smoothed_per_million'], label='Desenvolvidos')
ax_new_deaths.plot(df_brasil['date'], df_brasil['new_deaths_smoothed_per_million'], label='Brasil')
ax_new_deaths.set_title('Novas mortes a cada 1.000.000 de pessoas')
ax_new_deaths.legend()
````

# **3. Conclusão**

- O país apesar de um início conturbado, principalmente no primeiro ano de pandemia (2021) conseguiu conter o vírus após o vasto número de vacinas aplicadas em todo o território. Contudo, é de extrema importante ressaltar o grande número de mortes ocorridas no Brasil, principalmente devido a falta de infraestutura da saúde no país, no qual não conseguiu suportar o grande aumento de pacientes infectados.

- Além disso, apesar de parecer controlado o número de casos comparado com os países desenvolvidos, o Brasil aparenta ter apresentado valores que não refletem a realidade, principalmente pelo baixissimo nível de testes aplicados no período.
